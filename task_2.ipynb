# Exploratory Data Analysis(EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titan=pd.read_csv('Titanic-Dataset.csv')
```

```python
titan
```

```python
# summary statistics for the numerical data
titan.describe()
```

```python
# summary for categorical feature
titan.describe(include='object')
```

```python
titan['Family_size']=titan['SibSp'] + titan['Parch'] + 1
```

```python
# correlation heatmap
numeric_df=titan.select_dtypes(include=[np.number])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(),annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=numeric_df)
plt.title('Box Plot')
plt.show()
```

```python
sns.pairplot(data=titan)
```

```python
plt.figure(figsize=(12,8))
sns.histplot(data=titan['Fare'])
plt.show()
```

```python
plt.rc('figure',figsize=(12,8))
fig_size_subplot=(10,14)
bin_size=10
```

```python
fig=plt.figure(figsize=fig_size_subplot)
fig_dims=(3,2)
plt.subplot2grid(fig_dims,(0,0))
titan['Survived'].value_counts().plot(kind='bar',title='Death vs. Survival counts')

plt.subplot2grid(fig_dims,(0,1))
titan['Pclass'].value_counts().plot(kind='bar',title='Passenger Class counts')

plt.subplot2grid(fig_dims,(1,0))
titan['Sex'].value_counts().plot(kind='bar',title='Passenger Class counts')

plt.xticks(rotation=0)
plt.subplot2grid(fig_dims,(1,1))
titan['Embarked'].value_counts().plot(kind='bar',title='Embarked counts')

plt.subplot2grid(fig_dims,(2,0))
titan['Age'].hist()
plt.title('Age Histogram')
```

```python
sns.distplot(titan['Family_size'])
```
